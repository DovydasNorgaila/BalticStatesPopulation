Q3. Sukurkite Python žodyną (dictionary) su Baltijos šalių gyventojų skaičiumi

```python
df_baltic_population = {
    "Lietuva": 2.82,
    "Latvija": 1.94,
    "Estija": 1.30
}
print('Gyventojų skaičius, mln. :', df_baltic_population)
```

Q4. Sukurkite Python žodyną (dictionary) su Baltijos šalių sostinių gyventojų skaičiumi.

```python
df_baltic_capitalpopulation = {
    "Vilnius": 559.421,
    "Ryga": 614.618,
    "Talinas": 438.341
}
print('Gyventojų skaičius sostinėse, tūkst. :', df_baltic_capitalpopulation)
```

Q5. Pavaizduokite Baltijos šalių gyventojų skaičių stulpelinėjė diagramoje

```python
from matplotlib import pyplot as plt
import pandas as pd
%matplotlib inline

baltic_countries = df_baltic_population.keys()
df_baltic_population_list = df_baltic_population.values()

plt.figure(figsize=(14, 6))
bar = plt.bar(baltic_countries, df_baltic_population_list, color="blue")
plt.title("\nBaltijos šalių gyventojų skaičius, mln.\n", fontsize = 22)
plt.ylabel("Gyventojai, mln.", fontsize = 22)

plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)

# Bar labels
labels_population_in_mln = [f"{p} mln." for p in df_baltic_population_list]
plt.bar_label(bar, labels=labels_population_in_mln, padding=-30, fontsize=18, color="white")

plt.grid(axis="y")
plt.show()
```

Q6. Pavaizduokite Baltijos šalių sostinių gyventojų skaičių stulpelinėjė diagramoje.

```python
from matplotlib import pyplot as plt
import pandas as pd
%matplotlib inline

baltic_countries = df_baltic_capitalpopulation.keys()
df_baltic_capitalpopulation_list = df_baltic_capitalpopulation.values()

plt.figure(figsize=(14, 6))
bar = plt.bar(baltic_countries, df_baltic_capitalpopulation_list, color="orange")
plt.title("\nBaltijos šalių sostinių gyventojų skaičius, tūktst.\n", fontsize = 22)
plt.ylabel("Gyventojai, tūkst.", fontsize = 22)

plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)

# Bar labels
labels = [f"{round(p,1)} tūkst." for p in df_baltic_capitalpopulation_list]
plt.bar_label(bar, labels=labels, padding=-30, fontsize=18, color="white", fmt="%0.0f")

plt.grid(axis="y")
plt.show()
```

Q7. Pavaizduokite Baltijos šalių gyventojų proporcijas skritulinėje diagramoje

```python
plt.figure(figsize = (12, 6))

plt.pie(df_baltic_population_list, labels=baltic_countries, autopct = "%1.0f%%", textprops = {"fontsize": 14})

plt.title("\nBaltijos šalių gyventojų pasiskirstymas\n", fontsize = 22)
plt.show()
```

Q8. Suraskite Lietuvos, Latvijos ir Estijos gyventojų skaičiaus dinamiką nuo 2015 iki 2020. Pavaizduokite gyventojų skaičiaus keitimą linijinėje diagramoje.

```python
# df_baltic_population DataFrame (Q8, Q9)

Year = pd.Series(['2015', '2016', '2017', '2018', '2019', '2020'])
Estonia_population = pd.Series([1.315, 1.315, 1.317, 1.321, 1.326, 1.329])
Latvia_population = pd.Series([1.977, 1.977, 1.942, 1.927, 1.913, 1.900])
Lithuania_population = pd.Series([2.904, 2.868, 2.828, 2.801, 2.794, 2.794])

df_baltic_population = pd.DataFrame({ 'Year': Year, 'Estonia': Estonia_population, 'Latvia': Latvia_population, 'Lithuania': Lithuania_population})
df_baltic_population

df_baltic_population.set_index('Year', inplace=True)
```

```python
fig, ax = plt.subplots(figsize=(14,6))
df_baltic_population.plot(ax=ax, linewidth=3)

plt.title("\nBatltijos šalių gyventojų skaičiaus dinamika\n", fontsize = 22)
plt.xlabel("Metai", fontsize = 18)
plt.ylabel("Gyventojai, mln.", fontsize = 18)

plt.xticks(fontsize = 16)
plt.yticks(fontsize = 16)

plt.legend(loc="upper right", fontsize = 16, bbox_to_anchor=(1.3, 1))

plt.grid()
plt.show()
```
