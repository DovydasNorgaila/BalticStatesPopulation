Q5. Pavaizduokite Baltijos šalių gyventojų skaičių stulpelinėjė diagramoje

```python
from matplotlib import pyplot as plt
import pandas as pd
%matplotlib inline

baltic_countries = df_baltic_population.keys()
df_baltic_population_list = df_baltic_population.values()

plt.figure(figsize=(14, 6))
plt.bar(baltic_countries, df_baltic_population_list, color="blue")
plt.title("\nBaltijos šalių gyventojų skaičius, mln.\n", fontsize = 22)
plt.ylabel("Gyventojai, mln.", fontsize = 22)

plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)

plt.grid()
plt.show()
```

Q8. Suraskite Lietuvos, Latvijos ir Estijos gyventojų skaičiaus dinamiką nuo 2015 iki 2020. Pavaizduokite gyventojų skaičiaus keitimą linijinėje diagramoje.

```python
import pandas as pd
url = 'https://raw.githubusercontent.com/Dovis230/yo/main/baltic_countries.csv?token=GHSAT0AAAAAABT6GHG27DDJLNRTYXLNGTHMYTHW7TA'
df_baltic_population = pd.read_csv(url)
df_baltic_population
```

```python
Year = pd.Series(['2015', '2016', '2017', '2018', '2019', '2020'])
Estonia_population = pd.Series([1.315, 1.315, 1.317, 1.321, 1.326, 1.329])
Latvia_population = pd.Series([1.977, 1.977, 1.942, 1.927, 1.913, 1.900])
Lithuania_population = pd.Series([2.904, 2.868, 2.828, 2.801, 2.794, 2.794])

df_baltic_population = pd.DataFrame({ 'Year': Year, 'Estonia': Estonia_population, 'Latvia': Latvia_population, 'Lithuania': Lithuania_population})
df_baltic_population

df_baltic_population.set_index('Year')
```

```python
fig, ax = plt.subplots(figsize=(14,6))
df_baltic_population.plot(ax=ax)

plt.title("\nBatltijos šalių gyventojų skaičiaus dinamika\n", fontsize = 22)
plt.xlabel("Metai", fontsize = 18)
plt.ylabel("Gyventojai, mln.", fontsize = 18)

plt.xticks(fontsize = 16)
plt.yticks(fontsize = 16)

labels = [item.get_text() for item in ax.get_xticklabels()]
labels[1] = '2015'
labels[2] = '2016'
labels[3] = '2017'
labels[4] = '2018'
labels[5] = '2019'
labels[6] = '2020'

ax.set_xticklabels(labels)

plt.legend(loc="upper right", fontsize = 16, bbox_to_anchor=(1.3, 1))

plt.grid()
plt.show()
```
